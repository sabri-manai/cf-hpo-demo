```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/ready_for_surrogates.csv", low_memory=False)
df.head().T
```

```python
    columns_to_drop = [
                        "user_attrs_roc_auc_test", "user_attrs_roc_auc_val", 
                        "user_attrs_acc_val", "user_attrs_f1_test",
                        "user_attrs_f1_val", "user_attrs_pr_auc_test",
                        "user_attrs_pr_auc_val", "user_attrs_prec_test", 
                        "user_attrs_prec_val", "user_attrs_rec_test",
                        "user_attrs_rec_val", "user_attrs_acc_test"
    ]
    df_without_metrices = df.drop(columns_to_drop, axis=1)
    df_without_metrices.head().T
```

```python
bins = [0, 0.5, 0.6, 0.7, 1]
labels = ['very-low', 'low', 'moderate', 'high']
df_without_metrices["cat_value"] = pd.cut(df_without_metrices["value"], bins=bins, labels=labels, include_lowest=True)
df_without_metrices.head().T
```

```python
acfx_df = df_without_metrices.drop("value", axis=1)
acfx_df.head().T
```

```python
csv_path = "../data/ready_for_acfx.csv"
acfx_df.to_csv(csv_path, index=False)
print(f"\nSaved clean dataset to: {csv_path}  | shape={acfx_df.shape}")
```
